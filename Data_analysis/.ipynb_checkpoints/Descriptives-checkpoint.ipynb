### Word Count

```python
def descriptif(df):   
    df['num_words_cineuropa'] = df['cineuropa_review_text'].apply(lambda x: len(x.split()))
    df['num_words_variety'] = df['variety_review_text'].apply(lambda x:x if pd.isna(x) else len(x.split()))
    df['num_words_hollywood'] = df['hollywoodreporter_review_text'].apply(lambda x:x if pd.isna(x) else len(x.split()))
    df['num_words_screendaily'] = df['screendaily_review_text'].apply(lambda x:x if pd.isna(x) else len(x.split()))
descriptif(df) 
```

### Popular words

```python
 df['clean_cineuropa'] = df['cineuropa_review_text'].astype(str)
 df['clean_cineuropa'] =  df['clean_cineuropa'].str.replace('[^\w\s]','')

def create_text_blob(df, text_column):
    blob_text=[]
    for i, content in df[text_column].items():
        for i in content.split():
            blob_text.append(i.lower())
    return blob_text

blob_text = create_text_blob(df, 'clean_cineuropa')
print(blob_text[0:100])
```

```python
stop_words = set(stopwords.words('english'))  

filtered_sentence = [w for w in blob_text if not w in stop_words]  
filtered_sentence = []  
  
for w in blob_text:  
    if w not in stop_words:  
        filtered_sentence.append(w)  

print(filtered_sentence[0:100])  
```

```python
counts = Counter(filtered_sentence)
```

```python
import plotly.express as px

top_20_words = {}

for (key, value) in counts.items():
   # Check if value is greater than 200 and add to new dictionary
    if value > 1000 :
        top_20_words[key] = value
    continue

sorted_top_20_words = dict(sorted(top_20_words.items(), key=lambda item: item[1], reverse=False))

word = sorted_top_20_words.keys()
count = sorted_top_20_words.values()


fig = px.bar(y=word, x=count, text = count)
fig.update_traces(texttemplate='%{text:}', textposition='outside')
fig.update_layout(uniformtext_minsize=8, uniformtext_mode='hide')
fig.show()
```
